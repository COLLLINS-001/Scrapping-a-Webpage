```python
import requests
import re
from bs4 import BeautifulSoup
```

```python
response = requests.get('https://keithgalli.github.io/web-scraping/webpage.html')
web_content = response.content.decode('utf-8')
soup = BeautifulSoup(web_content, 'html.parser')
```

```python
SCRAPE THE TABLE
```

```python
table = soup.select('table')
```

```python
soup.select('table.hockey-stats')
```

```python
column = soup.find('thead').find_all('th')
columns = [c.string for c in column]
columns
```

```python
table_rows = soup.find('tbody').find_all('tr')
```

```python
import pandas as pd
l = []
for tr in table_rows:
    td = tr.find_all('td')
    row = [str(tr.text).strip() for tr in td]
    l.append(row)

df = pd.DataFrame(l, columns=columns)

df.head()
```

```python
soup.find_all('li', string = re.compile('is'))
```

```python
facts = soup.select('ul.fun-facts li')
facts_with_is = [fact.find(string = re.compile('is')) for fact in facts]
facts_with_is = [fact for fact in facts_with_is if fact]
facts_with_is
```
